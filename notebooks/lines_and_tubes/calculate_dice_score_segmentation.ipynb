```python
import numpy as np
import os
import pandas as pd
import torch
```

```python
def dice_score(x, y, thresholds=np.arange(0.1, 1.0, 0.9)):
    # assume x, y, are one-hot
    # x.shape = y.shape = (H, W)
    x = torch.stack([x >= _th for _th in thresholds])
    y = torch.stack([y] * len(thresholds))
    intersection = x * y
    denominator = x + y
    dice = 2 * intersection / denominator
    # dice.shape = (len(thresholds), )
    return dice
```
