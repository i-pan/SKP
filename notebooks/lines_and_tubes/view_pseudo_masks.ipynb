```python
import cv2
import glob
import matplotlib.pyplot as plt
import numpy as np

from skp.toolbox.functions import overlay_images
```

```python
files = glob.glob("/mnt/stor/datasets/kaggle/ranzcr-clip-catheter-line-classification/masks-pseudo-soft/*.png")
```

```python
x = np.random.choice(files)
mask = cv2.imread(x, cv2.IMREAD_UNCHANGED)[..., :3] # * 255
print(np.unique(mask))
img = cv2.imread(x.replace("masks-pseudo-soft", "train").replace("png", "jpg"), 0)
overlay = overlay_images(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB), mask)
plt.imshow(overlay)
plt.show()
```
